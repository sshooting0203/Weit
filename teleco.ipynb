```python
tele = tele.drop(columns='customerID')
```

```python
tele['TotalCharges'] = tele['TotalCharges'].replace(' ', np.nan) 
tele['TotalCharges'] = tele['TotalCharges'].astype(float)
tele.dropna(subset=['TotalCharges'], inplace=True)
```

```python
target = tele[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges','Churn']]
tele_encoded = pd.get_dummies(tele.drop(columns=['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges','Churn']))
tele_final = pd.concat([target, tele_encoded], axis=1)
tele_final.info()
```

```python
from sklearn.model_selection import train_test_split
tele_train,tele_test = train_test_split(tele_final,
                                        test_size=0.3,
                                        stratify=tele_final['Churn'],
                                        random_state=1234)
```

```python
tele_train.shape
```

```python
tele_test.shape
```

```python
tele_train['Churn'].value_counts(normalize=True)
```

```python
tele_test['Churn'].value_counts(normalize=True)
```

```python
from sklearn import tree
clf_3 = tree.DecisionTreeClassifier(random_state=1234,
                                  max_depth=3)
clf_4 = tree.DecisionTreeClassifier(random_state=1234,
                                  max_depth=4)
clf_5 = tree.DecisionTreeClassifier(random_state=1234,
                                  max_depth=5)
```

```python
train_x = tele_train.drop(columns='Churn')
train_y = tele_train['Churn']
model_3 = clf_3.fit(X = train_x, y = train_y)
model_4 = clf_4.fit(X = train_x, y = train_y)
model_5 = clf_5.fit(X = train_x, y = train_y)
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi':'100',
                     'figure.figsize':[12,8]})
tree.plot_tree(model_3,
              feature_names=list(train_x.columns),
              class_names=['No','Yes'],
              proportion = True,
              filled = True,
              rounded = True,
              impurity = False, #불순도 : 노드가 얼마나 혼합된 상태인지 측정하는 척도
              label='root',
              fontsize=10)
plt.show()
```

```python
test_x = tele_test.drop(columns='Churn')
test_y = tele_test['Churn']
```

```python
tele_test_3 = tele_test.copy()
tele_test_3['pred'] = model_3.predict(test_x)
tele_test_3
```

```python
tele_test_4 = tele_test.copy()
tele_test_4['pred'] =  model_4.predict(test_x)
tele_test_5 = tele_test.copy()
tele_test_5['pred'] =  model_5.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_true=tele_test_3['Churn'],
                            y_pred=tele_test_3['pred'],
                            labels=['No','Yes'])
conf_mat
```

```python
plt.rcParams.update(plt.rcParamsDefault)
from sklearn.metrics import ConfusionMatrixDisplay
p = ConfusionMatrixDisplay(confusion_matrix=conf_mat,
                           display_labels=('No','Yes'))
p.plot(cmap='Blues')
plt.show()
```

<p>행 : 실제 빈도(첫 번째 행 : 실제로 No인 사람, 두 번째 행 : 실제로 Yes인 사람)</p>
<p>열 : 모델이 예측한 빈도(첫 번째 열 : Yes로 예측한 사람, 두 번째 열 : No로 예측한 사람)</p>

<h3>정확도(Accuracy)</h3>
<p>예측해서 맞춘 비율</p>

```python
import sklearn.metrics as metrics
a_3 = metrics.accuracy_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'])
a_4 = metrics.accuracy_score(y_true=tele_test_4['Churn'],
                       y_pred=tele_test_4['pred'])
a_5 = metrics.accuracy_score(y_true=tele_test_5['Churn'],
                       y_pred=tele_test_5['pred'])
print(a_3, a_4, a_5)
```

<h3>정밀도(Precision)</h3>
<p>관심 클래스를 예측해서 맞춘 비율</p>

```python
pre_3_1 = metrics.precision_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'],
                       pos_label='Yes')
pre_3_2 = metrics.precision_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'],
                       pos_label='No')
print("Yes : ",pre_3_1,", No : ",pre_3_2)
#Yes가 가장 높은 Depth는 3,  No가 가장 높은 Depth는 5
```

<h3>재현율(Recall)</h3>
<p>실제 데이터에서 관심 클래스를 찾아낸 비율</p>

```python
rec_3_1 = metrics.recall_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'],
                       pos_label='Yes')
rec_3_2 = metrics.recall_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'],
                       pos_label='No')
print("Yes : ",rec_3_1,", No : ",rec_3_2)
#Yes가 가장 높은 Depth는 5, No가 가장 높은  Depth는 3
```

```python
rec_3_1 = metrics.recall_score(y_true=tele_test_5['Churn'],
                       y_pred=tele_test_5['pred'],
                       pos_label='Yes')
rec_3_2 = metrics.recall_score(y_true=tele_test_5['Churn'],
                       y_pred=tele_test_5['pred'],
                       pos_label='No')
print("Yes : ",rec_3_1,", No : ",rec_3_2)
#Yes가 가장 높은 Depth는 5, No가 가장 높은  Depth는 3
```

<h4>고객 이탈 예측이므로 최대한 이탈할 고객(Yes)를 찾아내는 것이 중요함</h4>

<h3>F1 score</h3>
<p>recall과 precision의 조화평균</p>

```python
f1_3_1 = metrics.f1_score(y_true=tele_test_5['Churn'],
                       y_pred=tele_test_5['pred'],
                       pos_label='Yes')
f1_3_2 = metrics.f1_score(y_true=tele_test_5['Churn'],
                       y_pred=tele_test_5['pred'],
                       pos_label='No')
print("Yes : ",f1_3_1,", No : ",f1_3_2)
#Yes가 가장 높은 Depth는 5, No가 가장 높은 Depth는 3
#조화평균이 높다 : 정밀도와 재현율이 균형하다 
#조화평균이 낮다 : 정밀도와 재현율이 불균형하다 
```

```python
f1_3_1 = metrics.f1_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'],
                       pos_label='Yes')
f1_3_2 = metrics.f1_score(y_true=tele_test_3['Churn'],
                       y_pred=tele_test_3['pred'],
                       pos_label='No')
print("Yes : ",f1_3_1,", No : ",f1_3_2)
#Yes가 가장 높은 Depth는 5, No가 가장 높은 Depth는 3
```

<h4>Yes를 예측하는 것이 더 중요하다면...</h4>
<h4>현재 훈련데이터에서 No클래스가 다수를 차지하므로 Yes와 No의 불균형 조정 필요</h4>

<h3>SMOTH 사용</h3>
<h4>소수 클래스의 인스턴스들 사이에서 새로운 데이터를 선형 보간(Interpolation)하여 생성</h4>
<h4>default : 소수 클래스의 샘플 수가 다수 클래스의 샘플 수와 동일</h4>

```python
from imblearn.over_sampling import SMOTE
#over-sampling : 소수 클래스를 중복하여 샘플링
#under-sampling : 다수 클래스를 줄여서 학습 
smote = SMOTE(random_state=42) #자동으로 소수 클래스를 인식하고 가상데이터를 생성
x_resampled, y_resampled = smote.fit_resample(train_x, train_y)
# x_resampled.shape : (7228, 45)
# train_x.shape : (4922, 45)
```

```python
train_y.value_counts(normalize=True)
```

```python
y_resampled.value_counts(normalize=True)
```

```python
rclf_3 = tree.DecisionTreeClassifier(random_state=1234,
                                  max_depth=3)
rmodel_3 = rclf_3.fit(x_resampled, y_resampled)
```

```python
plt.rcParams.update({'figure.dpi':'100',
                     'figure.figsize':[12,8]})
tree.plot_tree(rmodel_3,
              feature_names=list(train_x.columns),
              class_names=['No','Yes'],
              proportion = True,
              filled = True,
              rounded = True,
              impurity = False, #불순도 : 노드가 얼마나 혼합된 상태인지 측정하는 척도
              label='root',
              fontsize=10)
plt.show()
```

```python
rtele_test_3 = tele_test.copy()
rtele_test_3['pred'] =  rmodel_3.predict(test_x)
metrics.accuracy_score(y_true=rtele_test_3['Churn'],
                       y_pred=rtele_test_3['pred'])
```

<h3>재현율(Recall)</h3>
<p>이전 학습보다 depth=3에서 Yes에 대한 수치가 증가 </p>































































```python
rrec_3_1 = metrics.recall_score(y_true=rtele_test_3['Churn'],
                       y_pred=rtele_test_3['pred'],
                       pos_label='Yes')
rrec_3_2 = metrics.recall_score(y_true=rtele_test_3['Churn'],
                       y_pred=rtele_test_3['pred'],
                       pos_label='No')
print("Yes : ",rrec_3_1,", No : ",rrec_3_2)
# Yes :  0.3850267379679144 , No :  0.9335054874112331
```
